```python
# =============================================================================
# ZELLE 1 (Markdown) — Erklärung
# =============================================================================
"""
## 3.2 Berechnung des Kraftbeiwerts $c_N$ und der Normalkraft $N'$

Der Kraftbeiwert $c_N$ wird durch numerische Integration der $c_p$-Verteilung 
über die normierte Profiltiefe $x/c$ berechnet:

$$c_N = \int_0^1 \left( c_{p,\text{unten}} - c_{p,\text{oben}} \right) \, d\left(\frac{x}{c}\right)$$

Die $c_p$-Werte ergeben sich direkt aus den gemessenen Differenzdrücken:

$$c_p(i) = \frac{p_i - p_\infty}{p_{\infty,\text{dyn}}} = \frac{\Delta p_i}{q_\infty}$$

Da die Drucksensoren bereits Differenzdrücke $p_i - p_\infty$ liefern, gilt:
$$c_p(i) = \frac{\text{Messwert [Pa]}}{q_\infty}$$

Gemäß den Hinweisen der Versuchsbeschreibung werden folgende Ergänzungen vorgenommen:
1. Der erste Messpunkt der Unterseite (FU1) wird auf den Wert der Oberseite (FO1) gesetzt.
2. Im Spalt zwischen Flügel und Klappe wird der Druckverlauf linear interpoliert.
3. An der Hinterkante der Klappe wird $c_p = 0$ angenommen.

Die Normalkraft pro Tiefeneinheit $N'$ ergibt sich zu:
$$N' = c_N \cdot q_\infty \cdot c$$

mit der Profiltiefe $c$ und dem Staudruck $q_\infty = 400\,\text{Pa}$.
"""


# =============================================================================
# ZELLE 2 (Code) — Hilfsfunktion für c_N-Berechnung
# =============================================================================

def berechne_cp_und_cN(pressure_wing_suction, pressure_wing_pressure,
                        pressure_flap_suction, pressure_flap_pressure,
                        pos_wing_suction, pos_wing_pressure,
                        pos_flap_suction, pos_flap_pressure,
                        p_dyn_ref):
    """
    Berechnet c_p und c_N für eine Messkonfiguration.

    Parameters
    ----------
    pressure_wing_*  : np.ndarray  Differenzdrücke [Pa] für Flügel-Ober-/Unterseite
    pressure_flap_*  : np.ndarray  Differenzdrücke [Pa] für Klappen-Ober-/Unterseite
    pos_wing_*       : np.ndarray  x/c-Positionen Flügel
    pos_flap_*       : np.ndarray  x/c-Positionen Klappe
    p_dyn_ref        : float       Dynamischer Referenzdruck [Pa]

    Returns
    -------
    cN    : float   Kraftbeiwert
    x_o   : array  x/c-Positionen Oberseite (sortiert)
    cp_o  : array  c_p Oberseite
    x_u   : array  x/c-Positionen Unterseite (sortiert)
    cp_u  : array  c_p Unterseite
    """

    # --- c_p berechnen ---
    cp_wing_suction  = pressure_wing_suction  / p_dyn_ref   # Oberseite Flügel
    cp_wing_pressure = pressure_wing_pressure / p_dyn_ref   # Unterseite Flügel
    cp_flap_suction  = pressure_flap_suction  / p_dyn_ref   # Oberseite Klappe
    cp_flap_pressure = pressure_flap_pressure / p_dyn_ref   # Unterseite Klappe

    # --- Ergänzung 1: FU1 = FO1 (Staupunkt-Näherung) ---
    cp_wing_pressure[0] = cp_wing_suction[0]

    # --- Oberseite zusammensetzen (Flügel + Klappe), nach x/c sortieren ---
    x_o  = np.concatenate([pos_wing_suction, pos_flap_suction])
    cp_o = np.concatenate([cp_wing_suction,  cp_flap_suction])
    sort_o = np.argsort(x_o)
    x_o, cp_o = x_o[sort_o], cp_o[sort_o]

    # --- Unterseite zusammensetzen (Flügel + Klappe), nach x/c sortieren ---
    x_u  = np.concatenate([pos_wing_pressure, pos_flap_pressure])
    cp_u = np.concatenate([cp_wing_pressure,  cp_flap_pressure])
    sort_u = np.argsort(x_u)
    x_u, cp_u = x_u[sort_u], cp_u[sort_u]

    # --- Ergänzung 3: cp = 0 an Hinterkante (x/c = max) ---
    # Nur anhängen wenn nicht schon vorhanden
    x_end = max(x_o[-1], x_u[-1])
    if x_o[-1] < x_end:
        x_o  = np.append(x_o,  x_end)
        cp_o = np.append(cp_o, 0.0)
    if x_u[-1] < x_end:
        x_u  = np.append(x_u,  x_end)
        cp_u = np.append(cp_u, 0.0)

    # --- c_N: numerische Integration (Trapezregel) ---
    # c_N = integral(cp_unten) - integral(cp_oben)
    cN = trapezoid(cp_u, x_u) - trapezoid(cp_o, x_o)

    return cN, x_o, cp_o, x_u, cp_u


# Profiltiefe c aus Positionsdaten bestimmen (Abstand Nase bis Hinterkante in m)
# Aus der Positionsdatei: die maximale x/c-Position entspricht der Hinterkante
# Der absolute Wert c wird aus 'Abstand von der Nase (x) in mm' / (x/c) bestimmt
c_profil_mm = POS['Abstand von der Nase (x) in mm'].dropna().max()
xc_max      = POS['Positionsvektor FO x/c[-]'].dropna().max()
c_profil    = (c_profil_mm / xc_max) / 1000   # Profiltiefe in Meter
print(f"Profiltiefe c = {c_profil*1000:.1f} mm")


# =============================================================================
# ZELLE 3 (Code) — c_N und N' für Anstellwinkel-Variation
# =============================================================================

alpha_var       = [-5, 0, 5, 10, 15, 20, 25]
excel_files_alpha = [
    '2024-05-06_11-48_alpha-5_eta-2_Messdaten.xls',
    '2024-05-06_11-52_alpha0_eta-2_Messdaten.xls',
    '2024-05-06_11-54_alpha5_eta-2_Messdaten.xls',
    '2024-05-06_11-57_alpha10_eta-2_Messdaten.xls',
    '2024-05-06_11-59_alpha15_eta-2_Messdaten.xls',
    '2024-05-06_12-02_alpha20_eta-2_Messdaten.xls',
    '2024-05-06_12-04_alpha25_eta-2_Messdaten.xls'
]

cN_alpha = []
N_prime_alpha = []

for i, alpha in enumerate(alpha_var):
    file_path = os.path.join(folder_path, excel_files_alpha[i])
    dat_tmp = pd.read_excel(file_path, sheet_name='Tabelle1', header=10, nrows=48)
    modul_1_data = dat_tmp.iloc[:, [0, 1, 2]].dropna()
    modul_2_data = dat_tmp.iloc[:, [3, 4, 5]].dropna()

    pressure_wing_suction_side  = modul_2_data.iloc[:len(pos_wing_suction_side), 2].values.copy()
    pressure_wing_pressure_side = modul_1_data.iloc[:len(pos_wing_pressure_side), 2].values.copy()
    pressure_flap_suction_side  = modul_2_data.iloc[len(pos_wing_suction_side):
                                    len(pos_wing_suction_side) + len(pos_flap_suction_side), 2].values.copy()
    pressure_flap_pressure_side = modul_1_data.iloc[len(pos_wing_pressure_side):
                                    len(pos_wing_pressure_side) + len(pos_flap_pressure_side), 2].values.copy()

    cN, _, _, _, _ = berechne_cp_und_cN(
        pressure_wing_suction_side, pressure_wing_pressure_side,
        pressure_flap_suction_side, pressure_flap_pressure_side,
        pos_wing_suction_side, pos_wing_pressure_side,
        pos_flap_suction_side, pos_flap_pressure_side,
        p_dyn_ref
    )
    N_prime = cN * p_dyn_ref * c_profil   # N' = c_N * q_inf * c  [N/m]

    cN_alpha.append(cN)
    N_prime_alpha.append(N_prime)
    print(f"alpha = {alpha:+3d}°  →  c_N = {cN:.4f},  N' = {N_prime:.2f} N/m")


# =============================================================================
# ZELLE 4 (Code) — Plot: c_N und N' über Anstellwinkel alpha
# =============================================================================

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Linke Achse: c_N
axes[0].plot(alpha_var, cN_alpha, 'o-', color='steelblue', linewidth=2, markersize=7)
axes[0].set_xlabel(r'Anstellwinkel $\alpha$ [°]', fontsize=13)
axes[0].set_ylabel(r'Kraftbeiwert $c_N$ [-]', fontsize=13)
axes[0].set_title(r'$c_N$ als Funktion des Anstellwinkels $\alpha$', fontsize=13)
axes[0].grid(True)
axes[0].axhline(0, color='gray', linewidth=0.8, linestyle='--')

# Rechte Achse: N'
axes[1].plot(alpha_var, N_prime_alpha, 's-', color='darkorange', linewidth=2, markersize=7)
axes[1].set_xlabel(r'Anstellwinkel $\alpha$ [°]', fontsize=13)
axes[1].set_ylabel(r"Normalkraft $N'$ [N/m]", fontsize=13)
axes[1].set_title(r"$N'$ als Funktion des Anstellwinkels $\alpha$", fontsize=13)
axes[1].grid(True)
axes[1].axhline(0, color='gray', linewidth=0.8, linestyle='--')

plt.tight_layout()
plt.show()


# =============================================================================
# ZELLE 5 (Code) — c_N und N' für Klappenwinkel-Variation
# =============================================================================

eta_K_var       = [-2, 5, 10, 15, 20, 25]   # -2° = Referenz (aus Alpha-Variation, alpha=5°)
excel_files_eta_K = [
    '2024-05-06_11-54_alpha5_eta-2_Messdaten.xls',   # Referenz eta=-2 bei alpha=5
    '2024-05-06_12-11_alpha5_eta5_Messdaten.xls',
    '2024-05-06_12-13_alpha5_eta10_Messdaten.xls',
    '2024-05-06_12-14_alpha5_eta15_Messdaten.xls',
    '2024-05-06_12-16_alpha5_eta20_Messdaten.xls',
    '2024-05-06_12-18_alpha5_eta25_Messdaten.xls'
]

cN_eta = []
N_prime_eta = []

for i, eta in enumerate(eta_K_var):
    file_path = os.path.join(folder_path, excel_files_eta_K[i])
    dat_tmp = pd.read_excel(file_path, sheet_name='Tabelle1', header=10, nrows=48)
    modul_1_data = dat_tmp.iloc[:, [0, 1, 2]].dropna()
    modul_2_data = dat_tmp.iloc[:, [3, 4, 5]].dropna()

    pressure_wing_suction_side  = modul_2_data.iloc[:len(pos_wing_suction_side), 2].values.copy()
    pressure_wing_pressure_side = modul_1_data.iloc[:len(pos_wing_pressure_side), 2].values.copy()
    pressure_flap_suction_side  = modul_2_data.iloc[len(pos_wing_suction_side):
                                    len(pos_wing_suction_side) + len(pos_flap_suction_side), 2].values.copy()
    pressure_flap_pressure_side = modul_1_data.iloc[len(pos_wing_pressure_side):
                                    len(pos_wing_pressure_side) + len(pos_flap_pressure_side), 2].values.copy()

    cN, _, _, _, _ = berechne_cp_und_cN(
        pressure_wing_suction_side, pressure_wing_pressure_side,
        pressure_flap_suction_side, pressure_flap_pressure_side,
        pos_wing_suction_side, pos_wing_pressure_side,
        pos_flap_suction_side, pos_flap_pressure_side,
        p_dyn_ref
    )
    N_prime = cN * p_dyn_ref * c_profil

    cN_eta.append(cN)
    N_prime_eta.append(N_prime)
    print(f"eta_K = {eta:+3d}°  →  c_N = {cN:.4f},  N' = {N_prime:.2f} N/m")


# =============================================================================
# ZELLE 6 (Code) — Plot: c_N und N' über Klappenwinkel eta_K
# =============================================================================

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].plot(eta_K_var, cN_eta, 'o-', color='steelblue', linewidth=2, markersize=7)
axes[0].set_xlabel(r'Klappenwinkel $\eta_K$ [°]', fontsize=13)
axes[0].set_ylabel(r'Kraftbeiwert $c_N$ [-]', fontsize=13)
axes[0].set_title(r'$c_N$ als Funktion des Klappenwinkels $\eta_K$', fontsize=13)
axes[0].grid(True)

axes[1].plot(eta_K_var, N_prime_eta, 's-', color='darkorange', linewidth=2, markersize=7)
axes[1].set_xlabel(r'Klappenwinkel $\eta_K$ [°]', fontsize=13)
axes[1].set_ylabel(r"Normalkraft $N'$ [N/m]", fontsize=13)
axes[1].set_title(r"$N'$ als Funktion des Klappenwinkels $\eta_K$", fontsize=13)
axes[1].grid(True)

plt.tight_layout()
plt.show()
```
