$$ y = y_1 + \frac{(x - x_1)}{(x_2 - x_1)} \times (y_2 - y_1) $$

# Funktion zur Interpolation eines fehlenden Werts
Diese Funktion schätzt einen spezifischen fehlenden Wert in einem Array `y` durch den Mittelwert der benachbarten Werte.

`interpolate_missing_values_manually(y, index)`: Berechnet den fehlenden Wert an der Position `index` als den Mittelwert der Nachbarn:
  $$
  y[\text{index}] = \frac{y[\text{index} - 1] + y[\text{index} + 1]}{2}
  $$

````python
# Funktion zur Interpolation eines bestimmten fehlenden Werts
def interpolate_missing_values_manually(y, index):
    """ Interpoliert einen spezifischen fehlenden Wert durch den Mittelwert der Nachbarwerte """
    y[index] = (y[index - 1] + y[index + 1]) / 2
    return y
````

# Numerische Integration mit `scipy.trapezoid`

Die Funktion `scipy.integrate.trapezoid` implementiert die Trapezregel, um das Integral numerisch zu berechnen.

Die Trapezregel berechnet die Fläche eines Trapezes zwischen zwei Punkten $x_1$ und $x_2$ durch:

$$\int_a^b f(x) \, dx \approx \sum_{i=1}^{n-1} \frac{(x_{i+1} - x_i)}{2} \left( f(x_i) + f(x_{i+1}) \right)$$

## Beispiel in Python mit `scipy.integrate.trapezoid`:

